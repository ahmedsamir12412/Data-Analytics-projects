Next, let's handle the missing values. For simplicity, we can remove the rows with missing values. However, in a real-world scenario, a better strategy might be needed depending on the nature and amount of the missing data.

### Next, let's dive deeper into the data and extract more insights. We'll look at the distribution of car prices, the relationship between different features and the car price, the most common car types, etc.

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the visualization
sns.set(style='ticks')

# Create a figure and a set of subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(20, 15))

# Plot the distribution of car prices
sns.histplot(data['Price'], kde=True, ax=axs[0, 0])
axs[0, 0].set_title('Distribution of Car Prices')

# Plot the relationship between engine size and car price
sns.scatterplot(x='EngineSize', y='Price', data=data, ax=axs[0, 1])
axs[0, 1].set_title('Relationship between Engine Size and Car Price')

# Plot the relationship between horsepower and car price
sns.scatterplot(x='Horsepower', y='Price', data=data, ax=axs[1, 0])
axs[1, 0].set_title('Relationship between Horsepower and Car Price')

# Plot the relationship between weight and car price
sns.scatterplot(x='Weight', y='Price', data=data, ax=axs[1, 1])
axs[1, 1].set_title('Relationship between Weight and Car Price')

plt.tight_layout()
plt.show()
```

### Here are the visualizations based on the dataset:

* Distribution of Car Prices: This histogram shows the distribution of car prices in the dataset. Most of the cars are priced between 10,000 and 40,000.
* Relationship between Engine Size and Car Price: This scatter plot shows the relationship between the engine size and the car price. There seems to be a positive correlation between engine size and car price, indicating that cars with larger engines tend to be more expensive.
* Relationship between Horsepower and Car Price: This scatter plot shows the relationship between horsepower and car price. There is a positive correlation between horsepower and car price, indicating that cars with more horsepower tend to be more expensive.
* Relationship between Weight and Car Price: This scatter plot shows the relationship between the weight of the car and its price. There seems to be a positive correlation between weight and car price, indicating that heavier cars tend to be more expensive.

### These insights can be valuable for understanding the factors that influence car prices.

### Q: How many each car type?

```python
# Plot the count of each car type
plt.figure(figsize=(10, 6))
plot=sns.countplot(x='Type', data=data)
plt.title('Count of Each Car Type')

# Adding the text labels for each bar
for p in plot.patches:
    plot.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 10), 
                   textcoords = 'offset points')

plt.show()
```

From the chart, we can see that **'Sedan'** is the most common car type in the dataset, followed by **'SUV'** and **'Sports'**. **'Hybrid'** and **'Truck'** are the least common car types.

### This information can be useful for understanding the distribution of car types in the dataset.

### Next, let's look at the average price of each car type.

```python
# Calculate the average price of each car type
type_avg_price = data.groupby('Type')['Price'].mean()

# Plot the average price of each car type
plt.figure(figsize=(10, 8))
vis_avg_price=sns.barplot(x=type_avg_price.index, y=type_avg_price.values)
plt.title('Average Price of Each Car Type')
plt.xlabel('Car Type')
plt.ylabel('Average Price')

# Adding the text labels for each bar
for p in vis_avg_price.patches:
    vis_avg_price.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 10), 
                   textcoords = 'offset points')

plt.show()
```

### From the chart, we can see that 'Sports' cars have the highest average price, followed by 'SUV' and 'Wagon'. 'Truck' and 'Hybrid' cars have relatively lower average prices.

### This information can be useful for understanding the pricing trends of different car types in the dataset.

### Next, let's look at the average price of cars from each origin.

```python
# Plot a box plot for the relationship between car type and price
def plot_box_plot(data, x, y, title, xlabel=None, ylabel=None):
    plt.figure(figsize=(10,6))
    sns.boxplot(data=data, x=x, y=y)
    plt.title(title)
    if xlabel:
        plt.xlabel(xlabel)
    if ylabel:
        plt.ylabel(ylabel)
    plt.show()

plot_box_plot(data, 'Origin', 'Price', 'Relationship between Origin and Price')
```

### From the box plot, we can see that cars from Europe tend to be more expensive, while cars from Asia and the USA are less expensive. This could be due to European cars often being luxury vehicles with more features.

### Next, let's examine the relationship between car type and price. This could give us insight into how the type of a car affects its price.

```python
plot_box_plot(data, 'Type', 'Price', 'Relationship between Car Type and Price')
```

### From the box plot, we can see that the median price of sports cars is the highest among all car types. Sedans, SUVs, and wagons have similar median prices, while trucks have the lowest median price.

### Based on the insights we've gathered so far, let's continue our analysis by looking at the relationship between the number of cylinders in a car and its price. This could give us insight into how the number of cylinders in a car's engine affects its price.

```python
plot_box_plot(data, 'Cylinders', 'Price', 'Relationship between Number of Cylinders and Price')
```

### From the box plot, we can see that cars with more cylinders tend to be more expensive. This is expected as cars with more cylinders usually have more powerful engines, which can increase the price.

### Next, let's examine the relationship between drivetrain and price. This could give us insight into how the drivetrain of a car affects its price.


```python
plot_box_plot(data, 'DriveTrain', 'Price', 'Relationship between Drivetrain and Price')
```

**From the box plot, we can see that cars with rear-wheel drive tend to be more expensive than those with front-wheel or all-wheel drive. This could be due to the fact that rear-wheel drive is often found in high-performance and luxury cars, which are typically more expensive.**

**Next, let's examine the relationship between car length and price. This could give us insight into how the size of a car affects its price.**

```python
# Plot a scatter plot for the relationship between engine size and horsepower
def plot_scatter_plot(data, x, y, title, xlabel=None, ylabel=None):
    plt.figure(figsize=(10,6))
    sns.scatterplot(data=data, x=x, y=y)
    plt.title(title)
    if xlabel:
        plt.xlabel(xlabel)
    if ylabel:
        plt.ylabel(ylabel)
    plt.show()

plot_scatter_plot(data, 'Length', 'Price', 'Relationship between Car Length and Price')
```

**From the scatter plot, we can see a slight positive correlation between car length and price. This means that larger cars (in terms of length) tend to be slightly more expensive. This could be due to larger cars often having more features and being more comfortable, which can increase the price.**

**Next, let's examine the relationship between car weight and price. This could give us insight into how the weight of a car affects its price.**

```python
plot_scatter_plot(data, 'Weight', 'Price', 'Relationship between Car Weight and Price')
```

**From the scatter plot, we can see a positive correlation between car weight and price. This means that heavier cars tend to be more expensive. This could be due to heavier cars often being larger and having more features, which can increase the price.**

**Next, let's examine the relationship between city MPG and price. This could give us insight into how the fuel efficiency of a car in the city affects its price.**

```python
plot_scatter_plot(data, 'MPG_City', 'Price', 'Relationship between City MPG and Price')
```

**From the scatter plot, we can see a slight negative correlation between city MPG and price. This means that cars with higher fuel efficiency in the city tend to be slightly less expensive. This could be due to more fuel-efficient cars often being smaller and less powerful, which can decrease the price.**

**Next, let's examine the relationship between highway MPG and price. This could give us insight into how the fuel efficiency of a car on the highway affects its price.**

```python
plot_scatter_plot(data, 'MPG_Highway', 'Price', 'Relationship between Highway MPG and Price')
```

**From the scatter plot, we can see a slight negative correlation between highway MPG and price. This means that cars with higher fuel efficiency on the highway tend to be slightly less expensive. This is similar to the trend we observed with city MPG.**
